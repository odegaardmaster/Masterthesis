# PLOTS

```
n_c = [i for i in range(1,11)] # Number of clusters array
foo_nc = [np.nan for i in range(1,11)] # empty array
```

## Inverse metamodelling

```
plt.figure(figsize=(15,10))
for i,r2 in enumerate(SXinv.T):
  if i < 10:
    plt.plot(n_c,r2,'-*')
  else:
    plt.plot(n_c,r2,'-D')

    
plt.gca().set_prop_cycle(None)

for i,r2 in enumerate(SYinv.T):
  if i < 10:
    plt.plot(n_c,r2,'--*')
  else:
    plt.plot(n_c,r2,'--v')
    
for i in range(SXinv.shape[1]):
  if i < 10:
    plt.plot(n_c,nan_nc,'-*',label=names[i])
  else:
    plt.plot(n_c,nan_nc,'-D',label=names[i])
    
  
plt.plot(n_c,nan_nc,'--',c='black',label='SY')
plt.plot(n_c,nan_nc,'-',c='black',label='SX')

    
plt.xticks(np.arange(1, 11, step=1))
plt.ylim(0,1)
plt.xlim((1,11))
plt.ylabel('R2 score')
plt.xlabel('Number of clusters')
plt.legend()
plt.savefig("drive/My Drive/Figures/R2plots/PR_inverse")
```

## Classical metamodelling of aggregated features

```
plt.figure(figsize=(15,10))
plt.plot(n_c[:],np.mean(SXorg_ag,axis=1),'-*')
plt.plot(n_c[:],np.mean(SXsincos_ag,axis=1),'-*')
plt.plot(n_c[:],np.mean(SXso_ag,axis=1),'-*')

plt.gca().set_prop_cycle(None)

plt.plot(n_c[:],np.mean(SYorg_ag,axis=1),'--*')
plt.plot(n_c[:],np.mean(SYsincos_ag,axis=1),'--*')
plt.plot(n_c[:],np.mean(SYso_ag,axis=1),'--*')

plt.gca().set_prop_cycle(None)

plt.plot(n_c,foo_nc,label='without additional terms')
plt.plot(n_c,foo_nc,label='with sincos')
plt.plot(n_c,foo_nc,label='with second order interactions')

plt.plot(n_c,foo_nc,'--*',c='black',label='SY clustering basis')
plt.plot(n_c,foo_nc,'-*',c='black',label='SX clustering basis')


plt.xticks(np.arange(1, 11, step=1))
plt.ylim(0,1)
plt.xlim((1,11))
plt.ylabel('R2 score')
plt.xlabel('Number of clusters')
plt.legend()
plt.savefig("drive/My Drive/Figures/R2plots/PRmutliple_ag")
```

## Classical metamodelling

```
plt.figure(figsize=(15,10))


plt.plot(n_c[:],SXorg,'-*',c='blue')
plt.plot(n_c[:],SXso,'-*',c='red')
plt.plot(n_c[:],SXsincos,'-*',c='green')


plt.plot(n_c[:],SYorg,'--*',c='blue')
plt.plot(n_c[:],SYso,'--*',c='red')
plt.plot(n_c[:],SYsincos,'--*',c='green')

plt.plot(n_c,nan_nc,'blue',label="Without added terms")
plt.plot(n_c,nan_nc,'red',label="With second order terms")
plt.plot(n_c,nan_nc,'green',label="With sincos terms")

plt.plot(n_c,nan_nc,'--*',label='SY clustering basis',c='black')
plt.plot(n_c,nan_nc,'-*',label='SX clustering basis',c='black')


plt.xticks(np.arange(1, 11, step=1))
plt.ylim(0,1)
plt.xlim((1,11))
plt.ylabel('R2 score')
plt.xlabel('Number of clusters')
plt.legend()
plt.savefig("drive/My Drive/Figures/R2plots/PR_multiple")
```
