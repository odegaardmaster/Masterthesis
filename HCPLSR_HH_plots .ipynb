## Inverse metamodelling

### SX as clustering basis

```
plt.figure(figsize=(15,10))
plt.title('SX inverse')
for i,r2 in enumerate(SXinv.T):
  plt.plot([i for i in range(1,11)],r2,'-*',label=names[i])
plt.xticks(np.arange(1, 11, step=1))
plt.ylim(0,1)
plt.xlim((1,11))
plt.ylabel('R2 score')
plt.xlabel('Number of clusters')
plt.legend()
plt.savefig("drive/My Drive/Figures/R2plots/HHSX_inverse")
```

### SY as clustering basis

```
plt.figure(figsize=(15,10))
plt.title('SY inverse')
for i,r2 in enumerate(SYinv.T):
  plt.plot([i for i in range(1,11)],r2,'-*',label=names[i])
plt.xticks(np.arange(1, 11, step=1))
plt.ylim(0,1)
plt.xlim((1,11))
plt.ylabel('R2 score')
plt.xlabel('Number of clusters')
plt.legend()
plt.savefig("drive/My Drive/Figures/R2plots/HHSY_inverse")
```
